```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.tail()
```

```python
df=df.rename(columns={'per_capita_income_(US$)':'per_capita_income'})
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year,df['per_capita_income'], color='green',marker='.')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year,df['per_capita_income'], color='red',marker='.')
plt.plot(df.year,reg.predict(df[['year']]), color='green')
plt.title('Per capita Income vs Year (Linear Regression)')
plt.show()
```
